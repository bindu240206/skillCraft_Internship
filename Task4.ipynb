```python
import pandas as pd
```

```python
df = pd.read_excel("marketing_campaign.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Income'] = df['Income'].fillna(df['Income'].mean())
```

```python
df.duplicated().sum()
```

```python
df['Total_Spending'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
```

```python
df.groupby('Education')['Total_Spending'].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby('Education')['Total_Spending'].mean().plot(kind='bar')

plt.title('Average Spending by Education')
plt.xlabel('Education')
plt.ylabel('Average Spending')

plt.show()
```

```python
df['Response'].value_counts()
```

```python
df['Response'].value_counts().plot(kind='bar')

plt.title('Campaign Response')
plt.xlabel('Response')
plt.ylabel('Count')

plt.show()
```

```python
plt.scatter(df['Income'], df['Total_Spending'])

plt.title('Income vs Total Spending')
plt.xlabel('Income')
plt.ylabel('Total Spending')

plt.show()
```

# Business Insights Report – Marketing Campaign Analysis

## Objective

The goal of this analysis was to understand customer purchasing behavior and evaluate the effectiveness of marketing campaigns using Exploratory Data Analysis (EDA).

## Key Findings

* Customers with higher income tend to spend more on products.
* Customers with PhD and Graduation education levels showed higher average spending.
* Only a small percentage of customers responded positively to the marketing campaign.
* Most customers did not accept the campaign offer, indicating lower campaign engagement.

## Visual Analysis

* Bar charts showed spending differences across education groups.
* Campaign response charts highlighted low conversion rates.
* Scatter plots revealed a positive relationship between income and total spending.

## Business Recommendations

* Focus marketing budget on high-income customer segments.
* Create targeted campaigns for highly educated customers, as they spend more.
* Improve campaign personalization to increase customer response rates.
* Invest more in channels generating better customer engagement and conversions.

## Conclusion

EDA helped identify customer segments with higher spending potential and areas where marketing performance can be improved. Targeted campaigns and better customer segmentation can improve ROI and campaign effectiveness.
